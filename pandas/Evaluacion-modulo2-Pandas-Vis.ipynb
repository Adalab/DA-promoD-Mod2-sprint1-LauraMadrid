3. Preparación de los datos:
- Los nombres de las columnas empiezan con mayúsculas, poned todos los nombres de las columnas en minúsculas.
- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.
- Después de haber hecho los cambios, chequead los tipos de datos. ¿Son ya del tipo correcto? En caso de que no, cread otra función o modificar la anterior para que sean de tipo float.
- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "title" y "album".
- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.

```python
# cambiamos todos los nombres de las columnas para que empiecen con minúsculas

nuevas_columnas = {col: col.lower() for col in df_sp_yt.columns}
df_sp_yt.rename(columns = nuevas_columnas, inplace = True)
df_sp_yt.head(2)
```

```python
df_sp_yt.dtypes
```

```python
# creamos la función de cambio de datos

def cambio_tipo_dato(dato):
    """ esta fucion cabia las comas de los valores de las columnas que se le pase por puntos y cambia el tipo de datos a float. A la función
    se aplica el método apply para que haga la acción en todas las columnas que se le pase por parámetro"""
    
# ponemos el try/except caso haya valores nulos en las colunmas
    try:
        dato = float(dato.replace(",", "."))
        return dato
    except:
        return dato
```

```python

# creamos una lista con las columnas en las que iremos aplicar la función
lista_columnas = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]

# con el metodo .apply aplicamos la función a las columnas deseadas iterando por la lista de columnas 
for col in lista_columnas:
    df_sp_yt[col] = df_sp_yt[col] = df_sp_yt[col].apply(cambio_tipo_dato)
```

```python
df_sp_yt
```

```python
#info para probar que hemos cambiado los datos

df_sp_yt.info()
```

```python
# creamos la funcion que nos va a poner todas las columnas en minúsculas

def poner_columnasm (m):
    """ esta función cambia todas las columnas que se pasen por parámetro a columnas con minusculas"""

# ponemos el try/except caso haya valores nulos en las colunmas
    
    try:
        m = m.lower()
        return m
    except:
        return m
```

```python

# creamos una lista con las columnas en las que iremos aplicar la función
lista_columnasm = ["track", "title", "album"]

# con el metodo .apply aplicamos la función a las columnas deseadas iterando por la lista de columnas 
for col in lista_columnasm:
    df_sp_yt[col] = df_sp_yt[col].apply(poner_columnasm)
    
    
            
```

```python
df_sp_yt
```

```python
# creamos un patrón para quitar el símbolo de $ de los valores en la columna deseada

patron = "\$"
patron_n = "s"
```

```python
# creamos la funcion que nos va a  quitar en símbolo $ de los valoes de la columna deseada

def quitar_s(primero, segundo, df):
    
    
# con el try/except podemos ignorar los valores nulos 
    try:
        simbolo = re.sub(primero, segundo, df)
        return simbolo
    except:
        return np.NaN
```

```python
# con el metodo .apply aplicamos la función a las columnas deseadas iterando por la lista de columnas 

df_sp_yt["artist"] = df_sp_yt.apply(lambda dataframe: quitar_s(patron, patron_n, dataframe["artist"]), axis= 1)

df_sp_yt
```


4.3.  Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: 
- ¿Son aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis:
- Agrupar por artista y calcular la media de visitas y de likes.
- Quedaros con los 10 artistas que más visitas han tenido.
- Quedaros con los 10 artistas que más likes han tenido.
- Haced un subplot, donde en uno tengamos los 10 artistas con más likes y en el otro los 10 artistas con mas views.

```python
# sacamos los artistas con mas visitas y mas likes y calculamos la media

df_sp_yt_art = df_sp_yt.groupby("artist")[["views", "likes"]].mean()
df_sp_yt_art
```

```python
df_sp_yt_art.head()
```

```python

# sacamos los 10 primeros artistas con mas visitas 

df_sp_yt_views10 = df_sp_yt_art["views"].reset_index().sort_values(by="views", ascending=False).head(10)
df_sp_yt_views10
```

```python
# sacamos los 10 primeros artistas con mas likes

df_sp_yt_likes10 = df_sp_yt_art["likes"].reset_index().sort_values(by="likes", ascending=False).head(10)
df_sp_yt_likes10
```

```python
# creamos la gráfica con en número de visitas y likes de los top 10 artistas en spotify

fig,axes = plt.subplots(2,1, figsize = (12, 10))

# eligimos el barplot y establecemos las columnas y el df a utilizar 
sns.barplot(x = 'artist', 
            y = 'views',
            data = df_sp_yt_views10,
             ax = axes[0]) 

sns.barplot(x =  'artist',
            y = 'likes',
            data = df_sp_yt_likes10,
             ax = axes[1])
# quitamos las lineas de arriba y de la derecha
axes[0].spines[["top", "right"]].set_visible(False)
axes[1].spines[["top", "right"]].set_visible(False)

# ponemos en título general de la grafica y aumentamos su tamaño
fig.suptitle('Vistas y likes de los 10 top artistas en spotify', fontsize = 18)

# arreglamos la formatación de la gráfica 
plt.tight_layout()

plt.show()
```
